# #PLOTLY ASSIGNMENT

```python
#1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three
#dimensional space.

np.random.seed(30)
data = {'X': np.random.uniform(-10, 10, 300),
        'Y': np.random.uniform(-10, 10, 300),
        'Z': np.random.uniform(-10, 10, 300)}
df = pd.DataFrame(data)

fig=go.Figure()
fig.add_trace(go.Scatter3d(x=df.X, y=df.Y,z=df.Z,mode='markers',marker=dict(size=5,color=df.Z, colorscale='Viridis'
                          ,opacity=0.8)))
fig.update_layout(title='3D SCATTER PLOT OF RANDOM DATA' , scene=dict(xaxis_title='X axis',
                yaxis_title='Y axis', zaxis_title='Z axis'),width=800,height=700,margin=dict(r=20,b=10,l=10,t=40))
fig.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
'X': x.flatten(),
'Y': y.flatten(),
'Z': z.flatten()
}

df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(x=x,y=y,z=z)])
fig.update_layout(title='3D Surface plot',autosize=False,
width=800,height=800, margin=dict(l=65,r=50,b=65,t=90) ,         
scene=dict(xaxis_title='X',yaxis_title='Y', zaxis_title='Z'))

fig.show()
```

```python
#5. Using the given dataset, create a bubble chart to represent each country's population (y-axis),GDP (x
#axis), and bubble size proportional to the population.
np.random.seed(25)
data = {
'Country': ['USA', 'Canada', 'UK',
'Germany', 'France'],
'Population':
np.random.randint(100, 1000, 5),
'GDP': np.random.randint(500, 2000,
5)
}

df = pd.DataFrame(data)
fig=px.scatter(df, x='GDP',y='Population', 
               size='Population', title='Bubble Chart of population vs GDP',
               labels={'GDP':"GDP","Population":'Population'}, size_max=60)

fig.show()
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
from bokeh.models import LinearColorMapper, ColorBar
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10
#and y-values as the sine of x.

x=np.linspace(0,10,100)
y=np.sin(x)

p=figure(title='Sin Wave' , x_axis_label='X', y_axis_label='sin (x)')

p.line(x,y, legend_label='sin (x)') # line plot
show(p)
```
